Make sure to import "sklearn" to be able to retrieve this data-set:

<p style="color:purple">from sklearn.datasets import load_boston

```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 


boston = load_boston()
data=pd.DataFrame(boston.data,columns=boston.feature_names)

# Data

x = data["DIS"][:40].values.reshape(-1,1)
y = boston.target[:40]

xx=np.linspace(min(x),max(x),40)

from matplotlib import pyplot as plt
import numpy as np

# Linear Regression
ols = LinearRegression().fit(x,y)

plt.scatter(x,y)
plt.grid(True)
plt.title('Crime Vs Price', fontsize=14)
plt.plot(x,ols.predict(x),c='purple',lw=5)
plt.show()


# Polynomial Regression
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)
poly.fit(x_poly,y)
olp = LinearRegression().fit(x_poly,y)

plt.scatter(x,y)
plt.grid(True)
plt.title('Crime Vs Price', fontsize=14)
plt.plot(xx,olp.predict(poly.fit_transform(xx)),c='purple',lw=5)
plt.show()


```

<h2 style="color:red">Task 2:</h2>
<p>With the data-set of the task above do another polynomial regression by storing "AGE" as the x. The (y) doesn't change as it is our main data to compare with.

```python
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
data=pd.DataFrame(boston.data,columns=boston.feature_names)

x = data["AGE"][:40]
y = boston.target[:40]

from matplotlib import pyplot as plt
plt.scatter(x,y)

import numpy as np
yfit=np.poly1d(np.polyfit(x,y,2))
print(yfit)


xx=np.arange(min(x),max(x),0.1)
yy=[yfit(i) for i in xx]
plt.scatter(x,y)
plt.grid(True)
plt.title('Crime Vs Price', fontsize=14)
plt.plot(xx,yy,c='purple',lw=5)
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
data=pd.DataFrame(boston.data,columns=boston.feature_names)

x = data["DIS"][:40]
y = boston.target[:40]

from matplotlib import pyplot as plt
plt.scatter(x,y)

import numpy as np
yfit=np.poly1d(np.polyfit(x,y,2))
print(yfit)


xx=np.arange(min(x),max(x),0.1)
yy=[yfit(i) for i in xx]
plt.scatter(x,y)
plt.grid(True)
plt.title('Crime Vs Price', fontsize=14)
plt.plot(xx,yy,c='purple',lw=5)
plt.show()
```

```python
import pandas as pd

data=pd.read_csv('predict_admission.csv')

x = data["GRE Score"].iloc[:40]
y = data["Chance of Admit "].iloc[:40]

from matplotlib import pyplot as plt
plt.scatter(x,y)

import numpy as np

yfit=np.poly1d(np.polyfit(x,y,2))
print(yfit)
xx=np.arange(min(x),max(x),0.1)
yy=[yfit(i) for i in xx]
plt.scatter(x,y)
plt.grid(True)
plt.title('GRE Score Vs Chance of Admit', fontsize=14)
plt.plot(xx,yy,c='purple',lw=5)
plt.show()
```
